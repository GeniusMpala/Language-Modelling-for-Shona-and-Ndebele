```python
# Initialize the text-to-speech pipeline with your model

pipe = pipeline("text-to-speech", model="Fastino06/ff")

```

```python
# Provide the text you want to synthesize
text = "baba nevana zvakare"

# Generate the audio and save it to a file
audio = pipe(text)
with open("output.wav", "wb") as f:
  f.write(audio['audio']) # Access audio data using 'audio' key

# Optionally, you can also play the audio directly
from IPython.display import Audio
Audio(audio['audio'], rate=16000) # Provide the sampling rate
```
